Data Preprocessing

<span style="font-size:30px">
Data Preparation

```python
#Encodung categorical variable
sampled_data['gender'].unique()
```

```python
sampled_data['gender']= sampled_data['gender']. map({'Male': 0, 'Female': 1})
```

```python
sampled_data['family_history'].unique()
```

```python
sampled_data['family_history'] = sampled_data['family_history'].map({'Yes': 1, 'No': 0})
```

```python
sampled_data['smoking_status'].unique()
```

```python
sampled_data = pd.get_dummies(sampled_data, columns = ['smoking_status'], drop_first = True)
```

```python
smoking_columns = ['smoking_status_Former Smoker', 'smoking_status_Never Smoked', 'smoking_status_Passive Smoker']
sampled_data[smoking_columns] = sampled_data[smoking_columns].astype(int)
```

```python
sampled_data['cancer_stage'].unique()
```

```python
sampled_data = pd.get_dummies(sampled_data, columns = ['cancer_stage'], drop_first = True)
cancer_columns = ['cancer_stage_Stage II', 'cancer_stage_Stage III', 'cancer_stage_Stage IV']
sampled_data[cancer_columns] = sampled_data[cancer_columns].astype(int)
```

```python
sampled_data['treatment_type'].unique()
```

```python
sampled_data = pd.get_dummies(sampled_data, columns = ['treatment_type'], drop_first = True)
```

```python
treatment_columns = ['treatment_type_Combined', 'treatment_type_Radiation', 'treatment_type_Surgery']
sampled_data[treatment_columns] = sampled_data[treatment_columns].astype(int)
```

```python
sampled_data.head(5)
```

```python
#No missing data
pd.isnull(sampled_data).sum()
```

```python
#Dropping the diagnosis date, end treatment date, id and country as its not in use
sampled_data = sampled_data.drop(['diagnosis_date', 'end_treatment_date', 'country', 'id'], axis=1)

sampled_data.head()
```

```python
#splitting the data into training (70%), validation (15%), and test (15%) sets.
x = sampled_data[['age', 'gender','family_history','bmi', 'cholesterol_level', 'hypertension', 'asthma' , 'cirrhosis', 'other_cancer', 'smoking_status_Former Smoker', 'smoking_status_Never Smoked', 'smoking_status_Passive Smoker', 'cancer_stage_Stage II', 'cancer_stage_Stage III', 'cancer_stage_Stage IV', 'treatment_type_Combined', 'treatment_type_Radiation', 'treatment_type_Surgery']]
y = sampled_data['survived']

from sklearn.model_selection import train_test_split
#Spliting 70% train and 30% remaining
x_train, x_remaining, y_train, y_remaining = train_test_split(x, y, test_size=0.30, random_state=42, stratify = y)

#spliting remaining into 15% validation and 15% test
x_val, x_test, y_val, y_test = train_test_split(x_remaining, y_remaining, test_size=0.50, random_state=42, stratify=y_remaining)
```

```python
#Standardizing the numerical features
from sklearn.preprocessing import StandardScaler
numeric_columns = ['age','bmi', 'cholesterol_level']
standardization = StandardScaler()

x_train[numeric_columns] = standardization.fit_transform(x_train[numeric_columns])
x_val[numeric_columns]   = standardization.transform(x_val[numeric_columns])
x_test[numeric_columns]  = standardization.transform(x_test[numeric_columns])


```

```python
x_train[numeric_columns].head()
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
x_train_balanced, y_train_balanced = sm.fit_resample(x_train, y_train)
```

<span style="font-size:40px">
Supervised Machine Learning

<span style="font-size:30px">


<span style="font-size:30px">
Model Building - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 1000)
#training the model
model.fit(x_train_balanced, y_train_balanced)

#predicting using validation set
prediction_log_1 = model.predict(x_val)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
def evaluate_model(actual_value, predicted_value):
    print('Accuracy:', accuracy_score(actual_value, predicted_value))
    print('Precision:',precision_score(actual_value, predicted_value))
    print('Recall:',recall_score(actual_value, predicted_value))
    print('F1_score:',f1_score(actual_value, predicted_value))
    print('confusion_matrix:',confusion_matrix(actual_value, predicted_value))
    print('Classification:',classification_report(actual_value, predicted_value))

```

```python
evaluate_model(y_val, prediction_log_1)
```

```python
prediction_log_2 = model.predict(x_test)
evaluate_model(y_test, prediction_log_2)
```

<span style="font-size:30px">
Model Building - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(random_state = 42)
#training the model
model2.fit(x_train_balanced, y_train_balanced)

#Training Random Forest with GridSearchCV
from sklearn.model_selection import GridSearchCV
grid_params = {'n_estimators': [100, 200], 'max_depth': [None, 10, 20]}
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid=grid_params, cv=3, scoring='f1')
grid_search.fit(x_train_balanced, y_train_balanced)
model2 = grid_search.best_estimator_
#predicting using validation set
prediction_random_1 = model2.predict(x_val)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
def evaluate_model(actual_value, predicted_value):
    print('Accuracy:', accuracy_score(actual_value, predicted_value))
    print('Precision:',precision_score(actual_value, predicted_value))
    print('Recall:',recall_score(actual_value, predicted_value))
    print('F1_score:',f1_score(actual_value, predicted_value))
    print('confusion_matrix:',confusion_matrix(actual_value, predicted_value))
    print('Classification:',classification_report(actual_value, predicted_value))

```

```python
evaluate_model(y_val, prediction_random_1)
```

```python
#Prediction and Evaluation on Test Set
prediction_random_2 = model2.predict(x_test)
evaluate_model(y_test, prediction_random_2)
```

<span style="font-size:30px">
ROC/AUC ANALYSIS

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

logistic_probability = model.predict_proba(x_test)[:, 1]
random_forest_probability = model2.predict_proba(x_test)[:, 1]

logistic_fpr, logistic_tpr, _ = roc_curve(y_test, logistic_probability)
random_forest_fpr, random_forest_tpr, _ = roc_curve(y_test, random_forest_probability)

logistic_auc = auc(logistic_fpr, logistic_tpr)
random_forest_auc = auc(random_forest_fpr, random_forest_tpr)
```

```python
plt.figure(figsize = (8,6))
plt.plot(logistic_fpr, logistic_tpr, label = f'Logistic Regression(AUC = {logistic_auc:.4f})')
plt.plot(random_forest_fpr, random_forest_tpr, label=f'Random Forest(AUC = {random_forest_auc:.4f})', color = 'green')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve: survived or not')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate AUC scores
logistic_auc = auc(logistic_fpr, logistic_tpr)
random_forest_auc = auc(random_forest_fpr, random_forest_tpr)

print(f"Logistic Regression AUC: {logistic_auc:.4f}")
print(f"Random Forest AUC: {random_forest_auc:.4f}")
```

K Means

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#Standardizing the numerical columns
features_for_clustering = x.copy()
numeric_columns = ['age', 'bmi', 'cholesterol_level']
scaler = StandardScaler()
features_for_clustering[numeric_columns] = scaler.fit_transform(features_for_clustering[numeric_columns])

k_values = [3,4,5]
kmeans_models = {}
for k in k_values:
    kmeans = KMeans(n_clusters = k,random_state = 42, n_init =10)
    cluster_labels = kmeans.fit_predict(features_for_clustering)
    kmeans_models[k] = cluster_labels
    print(f"\nK={k} - Cluster Counts:", pd.Series(cluster_labels).value_counts())
```

```python
# b. Use PCA to reduce to 2D and visualize clusters (for k=3 as example)
pca = PCA(n_components=2)
features_2d = pca.fit_transform(features_for_clustering)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=features_2d[:, 0], y=features_2d[:, 1], hue=kmeans_models[3], palette='Set2')
plt.title("KMeans Clusters (k=3) Visualized in 2D using PCA")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Cluster Analysis
data_with_clusters = sampled_data.copy()
data_with_clusters['Cluster'] = kmeans_models[3]
cluster_analysis = data_with_clusters.groupby('Cluster').agg({
    'age': 'mean',
    'bmi': 'mean',
    'cholesterol_level': 'mean',
    'hypertension': 'mean',
    'asthma': 'mean',
    'cirrhosis': 'mean',
    'other_cancer': 'mean',
    'survived': 'mean'
})
print("\nCluster-wise Average Metrics:")
print(cluster_analysis)
```

ensemble

```python
from sklearn.ensemble import VotingClassifier, StackingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

lr = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
rf = RandomForestClassifier(random_state=42,  class_weight='balanced')
```

```python
#using stacking classifier
stacking_clf = StackingClassifier(estimators=[
    ('lr', lr),
    ('rf', rf)
], final_estimator=LogisticRegression(), passthrough=True)

stacking_clf.fit(x_train_balanced, y_train_balanced)
stacking_preds = stacking_clf.predict(x_test)
stacking_probs = stacking_clf.predict_proba(x_test)[:, 1]

def evaluate(name, y_true, y_pred, y_prob):
    print(f"\n--- {name} ---")
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall   :", recall_score(y_true, y_pred))
    print("F1 Score :", f1_score(y_true, y_pred))
    print("AUC Score:", roc_auc_score(y_true, y_prob))

evaluate("Stacking Classifier", y_test, stacking_preds, stacking_probs)
```

```python
#using voting classifier
voting_clf = VotingClassifier(estimators=[
    ('lr', lr),
    ('rf', rf)
], voting='soft')

voting_clf.fit(x_train_balanced, y_train_balanced)
voting_preds = voting_clf.predict(x_test)
voting_probs = voting_clf.predict_proba(x_test)[:, 1]

def evaluate(name, y_true, y_pred, y_prob):
    print(f"\n--- {name} ---")
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall   :", recall_score(y_true, y_pred))
    print("F1 Score :", f1_score(y_true, y_pred))
    print("AUC Score:", roc_auc_score(y_true, y_prob))

evaluate("Voting Classifier", y_test, voting_preds, voting_probs)
```
